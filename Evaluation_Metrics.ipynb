```python
## For evaluation : lebelled data

------------ Actual value ---
This is good movie                   Positive
This is bad movie                    Negative

----------- Predicted value --------> Predicted by model
This is good movie                   --
This is bad movie                    --
```

```python
## Confusion Metrics :

                        Predicted Positive                Predicted negative
Actual positive              TP                                  FN

Actual Negative              FP                                  TN
```

```python
## Accuracy : How many samples are correctly classified
Accuracy = (TP + TN) / TP + TN + FP + FN

## Can not trust accuracy when data is imbalance
```

```python
## Precision : What proportional of total predicted positives are actual positives
Precision = TP / (TP + FP)
```

```python
## Recall : what proportional of actual positive are correctly predicted
Recall = TP / (TP + FN)
```

```python
## F1-Score : Harmonic mean of precision & Recall
F1-Score = 2 * Precision * Recall / (Precision + Recall)
```
